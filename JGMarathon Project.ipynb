That was quick, now time to put it all together! This function iterates through all pages for each Marathon, taking in url and page count as arguements. The year and race arguements create column values for the dataframes

```python
def concatenator(base,pages, year, race):
    """Imports unique url's based on page number. Concatenates all dataframes into one df."""

    df_list = []
    
    
    for i in range(1, pages+1):
        url = (base + str(i))
    
        result = requests.post(url)
        soup = BS(result.text, 'lxml')
        tables = soup.find('table', attrs = {'class':"table table-responsive table-bordered"})
        results_list = pd.read_html(str(tables))
        df_list.append(results_list[0])
        
        df = pd.concat(df_list)
        df = df.reset_index()
        df["Year"] = str(year)
        df['Type'] = str(race)
        
        
    return df
   

        
```
